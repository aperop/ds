```python
import pandas as pd
```

## download and read the CSV file and make ID the index column

```python
dataset_link = "https://drive.google.com/open?id=1kFMUiXtrlw5B6WTjJ-mUm4-STV28Zsvn"
file_id = dataset_link.split("=")[-1]
!gdown {file_id} -q
!ls
```

```python
file_name = "auto.csv"
```

```python
df = pd.read_csv(file_name, index_col='ID')
df
```

## count the number of observations using the method count()

```python
df.count()
```

## drop the duplicates, taking into account only the following columns: CarNumber, Make_n_model, Fines

```python
df.drop_duplicates(subset=['CarNumber', 'Make_n_model', 'Fines'], keep='last', inplace=True)
```

```python
df.count()
```

## work with missing values

```python
df.isna().sum()
```

```python
df.dropna(axis=1, thresh=500, inplace=True)
df.isna().sum()
```

```python
df
```

```python
df.Refund.fillna(method='ffill', inplace=True)
df
```

```python
df.isna().sum()
```

```python
df.Fines.fillna(df.Fines.mean(), inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

## split and parse the make and model

```python
try:
    df.Make_n_model = df.Make_n_model.apply(str.split)
except:
    pass
df['Make'] = df.Make_n_model.apply(lambda x: x[0])
df['Model'] = df.Make_n_model.apply(lambda x: x[1] if len(x) > 1 else pd.NA)
```

```python
df
```

```python
df.isna().sum()
```

```python
df.drop(labels=['Make_n_model'], axis=1, inplace=True)
```

```python
df
```

```python
new_file_name = 'auto.json'
```

```python
df.to_json(new_file_name, orient='records')
```
